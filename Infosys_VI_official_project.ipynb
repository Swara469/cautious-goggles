```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("GlobalWeatherRepository.csv")
```

```python
print("First 5 rows:")
display(df.head())

print("\nDataset Info:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
df_clean = df.dropna()
df_clean.replace(-9999, np.nan, inplace=True)
df_clean.loc[df_clean['air_quality_PM10'] < 0, 'air_quality_PM10'] = np.nan
```

```python
print(df_clean.columns)
print("\nStatistical Summary:")
display(df_clean.describe())
```

```python
print("Duplicate Rows:", df_clean.duplicated().sum())
df_clean = df_clean.drop_duplicates()
```

```python
df_clean['last_updated'] = pd.to_datetime(df_clean['last_updated'])
monthly_avg = df_clean.resample(
    'ME',
    on='last_updated'
).mean(numeric_only=True).reset_index()

display(monthly_avg.head())
```

```python
df_clean.to_parquet("cleaned_weather_data.parquet", index=False)
```

```python
df_clean.to_csv("cleaned_weather_data.parquet", index=False)
monthly_avg.to_csv("monthly_average_weather.csv", index=False)

print("Files saved successfully!")
```

Extreme weather events: hottest days and wind speeds

```python
threshold = df_clean['temperature_celsius'].quantile(0.95)

extreme_heat = df_clean[df_clean['temperature_celsius'] > threshold]
extreme_heat.head()
```

```python
wind_threshold = df_clean['wind_kph'].quantile(0.95)
extreme_wind = df_clean[df_clean['wind_kph'] > wind_threshold]
extreme_wind.head()
```

Comparing weather across regions

```python
country_temp = df_clean.groupby('country')['temperature_celsius'].mean().sort_values(ascending=False)
fig = px.bar(
    country_temp.head(10),
    title="Top 10 Hottest Countries (Average Temperature)"
)
fig.show()
```

```python
import pandas as pd

# Make sure placeholders are cleaned
df_clean.replace(-9999, np.nan, inplace=True)

# Country-wise average temperature
country_temp = (
    df_clean.groupby('country')['temperature_celsius']
    .mean()
    .reset_index()
)

country_temp.head()
import plotly.express as px

fig = px.choropleth(
    country_temp,
    locations='country',
    locationmode='country names',  # IMPORTANT
    color='temperature_celsius',
    color_continuous_scale='RdYlBu_r',
    title='Average Temperature by Country (°C)'
)

fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True),
)

fig.show()
```

```python
country_pm25 = (
    df_clean.groupby('country')['air_quality_PM2.5']
    .mean()
    .reset_index()
)

fig = px.choropleth(
    country_pm25,
    locations='country',
    locationmode='country names',
    color='air_quality_PM2.5',
    color_continuous_scale='Reds',
    title='Average PM2.5 Levels by Country'
)

fig.show()
```
